```python
import copy

#Read through the qe output. Return a dictionary of raw qe info
def read_qe_output(filename):

    result = {
        'job_done':False,
        'total_energy': False,
        'beef_ensemble':False,
        'atom_dict':False,
    }

    #print(f"Searching for {filename}")
    file_found = True
    #print(f"file_found: {file_found}")

    if file_found == True:

        #print(f"The file '{filename}' has been found! Huzzah!")

        job_complete = False
        total_energy_found = False
        total_energy = 0

        line_nums_atoms = [0,0]
        line_nums_beef_ensemble = [0,0]

        unit_type = ""
        ry_to_eV = 13.6057039763


        with open(filename) as file:

            line_count = 0

            line_info_cart_axes = 'Cartesian axes'
            line_info_crystal_axes = 'Crystallographic axes'
            line_info_beef_ens_header = 'BEEFens 2000 ensemble energies'
            line_info_min_energy = '!    total energy'
            line_info_job_done = 'JOB DONE.'

            #Loop through every line in the output file
            for line in file:

                line_count += 1

                #1. Search for minimum energy
                if line_info_min_energy in line:
                    total_energy_found = True

                    unit = line.split(' ')[-1]
                    #self.debug_obj.debug_log(f"Energy unit: {unit}")

                    total_energy = float(line.split(" ")[-2])
                    #self.debug_obj.debug_log(f"Initial energy value: {total_energy}")

                    if "Ry" in unit:
                        #self.debug_obj.debug_log(f"Energy in Ry!")
                        unit_type = "Ry"
                        total_energy = total_energy*ry_to_eV

                #2. Search for jobe done
                if line_info_job_done in line:
                    job_complete = True
                    #print(f"Job complete found!")

                #3. Search for 'Cartesian axes' header
                if line_info_cart_axes in line:
                    line_nums_atoms[0] = line_count + 3

                #4. Search for 'Crystallographic axes' header
                if line_info_crystal_axes in line:
                    line_nums_atoms[1] = line_count - 2

                #5. Search for 'BEEFens 2000 ensemble energies'
                if line_info_beef_ens_header in line:
                    line_nums_beef_ensemble[0] = line_count + 1
                    line_nums_beef_ensemble[1] = line_count + 2000

            if total_energy_found and job_complete:
                result['job_done'] = True
                result['total_energy'] = total_energy
                #print(f"Total energy = {total_energy} eV!")
            else:
                return result

        #print(f"line_nums_atoms: {line_nums_atoms}")
        #print(f"line_nums_beef_ensemble: {line_nums_beef_ensemble}")

        #Now grab the atoms numbers and beef ensemble
        with open(filename) as file:

            line_count = 0

            atom_dict = {}
            beef_ens = []

            for line in file:
                line_count += 1

                if line_nums_atoms[0] <= line_count <= line_nums_atoms[1]:
                    atom_symbol = line.split()[1]
                    if atom_symbol in atom_dict:
                        atom_dict[atom_symbol] += 1
                    else:
                        atom_dict[atom_symbol] = 1

                if line_nums_beef_ensemble[0] <= line_count <= line_nums_beef_ensemble[1]:
                    raw_energy = float(line.split()[0])
                    clean_energy = 0
                    if unit_type == "Ry":
                        clean_energy = raw_energy*ry_to_eV
                    else:
                        clean_energy = raw_energy
                    beef_ens.append(clean_energy)

                if line_count > line_nums_beef_ensemble[1]:
                    break

            result['beef_ensemble'] = copy.deepcopy(beef_ens)
            result['atom_dict'] = copy.deepcopy(atom_dict)

    return result
```

```python
# Generate the 2d array of all meaningful data
import datetime

#Create 2d arrau
table_order = ['defect_name','beef_idx','optimal','centroid','relaxation','relaxation_all','charge','atoms','atoms_per_side','location','dfe','raw_energy','chem_pot','cube_root_of_volume_wrt_min_bond']

all_data = []

add_optimal = True
add_BEEF = True
experiment_filename = "PbI2_all_data.csv"

total_systems = len(defect_systems)
count = 0

for defect_system in defect_systems:

    count += 1
    print(f"Defect System: {defect_system} -- [{count}/{total_systems}]")
    start = datetime.datetime.now()

    defect_composition = defect_systems[defect_system]['defect_composition']

    pristine_result = read_qe_output(defect_systems[defect_system]["pristine_pwo"])
    defect_result = read_qe_output(defect_systems[defect_system]["defect_pwo"])

    pristine_number_of_atoms = pristine_result['atom_dict']['Pb']
    pristine_number_of_atoms += pristine_result['atom_dict']['I']

    pristine_ase = ase_read(defect_systems[defect_system]["pristine_pwo"])
    pristine_pymatgen = AseAtomsAdaptor.get_structure(atoms = pristine_ase)
    cube_root_of_volume = ((pristine_pymatgen.volume)**(1/3)) #pymat api gives volume in Angstrom^3
    cube_root_of_volume_wrt_min_bond = cube_root_of_volume/(defect_systems[defect_system]['pristine_min_bond_length'])

    entry_template = np.zeros(len(table_order),dtype=object)
    entry_template[table_order.index('defect_name')] = defect_systems[defect_system]['defect_name']
    entry_template[table_order.index('charge')] = defect_systems[defect_system]['charge']
    entry_template[table_order.index('atoms')] = defect_systems[defect_system]['atoms']
    entry_template[table_order.index('atoms_per_side')] = (pristine_number_of_atoms)**(1/3)
    entry_template[table_order.index('location')] = defect_systems[defect_system]['location']
    entry_template[table_order.index('centroid')] = False
    entry_template[table_order.index('optimal')] = False
    entry_template[table_order.index('cube_root_of_volume_wrt_min_bond')] = cube_root_of_volume_wrt_min_bond

    #Relaxation (Angstroms, NOT normalized)
    entry_template[table_order.index('relaxation')] = defect_systems[defect_system]['relaxation']
    if(defect_systems[defect_system]['relaxation_all'] == False):
        entry_template[table_order.index('relaxation_all')] = defect_systems[defect_system]['relaxation']
    else:
        entry_template[table_order.index('relaxation_all')] = defect_systems[defect_system]['relaxation_all']


    if(add_optimal):
        #First iterate through the 'optimal' chemical potential
        entry_template[table_order.index('optimal')] = True
        entry_template[table_order.index('beef_idx')] = 2000

        raw_energy = defect_result['total_energy'] - pristine_result['total_energy']
        entry_template[table_order.index('raw_energy')] = raw_energy
        raw_energy += defect_systems[defect_system]['FNV_correction']

        for chem_pot_coord in chemical_potential_optimal:
            #print(f"Chem Pot coordinate: {chem_pot_coord}")

            new_row = copy.copy(entry_template)

            #Calculate and add defect formation energy
            #print(f"Chemical Potenital")
            chem_energy = 0
            for species in defect_composition:
                chem_energy += (chem_pot_coord[species]*defect_composition[species])
            new_row[table_order.index('chem_pot')] = chem_energy
            new_row[table_order.index('raw_energy')] = raw_energy

            electron_energy_correction = defect_systems[defect_system]['charge'] * defect_systems[defect_system]['VBM']

            dfe = raw_energy - chem_energy + electron_energy_correction
            new_row[table_order.index('dfe')] = dfe

            #Update centroid
            if(chem_pot_coord['centroid'] == True):
                new_row[table_order.index('centroid')] = True

            all_data.append(new_row)

    if(add_BEEF):
        #Next iterate through the BEEF ensemble
        entry_template[table_order.index('optimal')] = False

        #Next iterate through the BEEF ensemble
        for beef_idx in range(0,2000):
            chem_pot_of_beef_idx = chemical_potential_BEEF[beef_idx]
            if not chem_pot_of_beef_idx:
                #print(f"Beef #{beef_idx} is empty.")
                continue

            entry_template[table_order.index('beef_idx')] = beef_idx

            pristine_energy = (pristine_result['total_energy'] + pristine_result['beef_ensemble'][beef_idx])
            defect_energy = (defect_result['total_energy'] + defect_result['beef_ensemble'][beef_idx])
            raw_energy = (defect_energy - pristine_energy)
            entry_template[table_order.index('raw_energy')] = raw_energy
            raw_energy += defect_systems[defect_system]['FNV_correction']

            for chem_pot_coord in chem_pot_of_beef_idx:
                new_row = copy.copy(entry_template)

                #Calculate and add defect formation energy
                chem_energy = 0

                for species in defect_composition:
                    chem_energy += (chem_pot_coord[species]*defect_composition[species])
                new_row[table_order.index('chem_pot')] = chem_energy

                electron_energy_correction = defect_systems[defect_system]['charge'] * defect_systems[defect_system]['VBM']

                dfe = raw_energy - chem_energy + electron_energy_correction
                new_row[table_order.index('dfe')] = dfe

                #Update centroid
                if(chem_pot_coord['centroid'] == True):
                    new_row[table_order.index('centroid')] = True


                all_data.append(new_row)

    stop = datetime.datetime.now()
    print(f"Duration: {stop-start}")

os.chdir(f'{beef_charged_defect_dir}/data_files/PbI2_mp22893/analysis')

import pandas as pd
pd_all_data = pd.DataFrame(all_data)
pd_all_data.columns = table_order
pd_all_data.to_csv(experiment_filename)
```
